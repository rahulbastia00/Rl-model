```python
import pandas as pd
df = pd.read_csv("filename.csv")
df
```

```python
import gymnasium as gym
import numpy as np
import pandas as pd

class CodeEfficiencyEnv(gym.Env):
    def __init__(self, df):
        super(CodeEfficiencyEnv, self).__init__()
        self.df = df
        self.index = 0
        
        # Define Action Space (0: Keep Code, 1: Suggest Optimization)
        self.action_space = gym.spaces.Discrete(2)

        # Define Observation Space (CPU_Time, Memory, Energy, Carbon_Emission, Loops, Control_Flows)
        self.observation_space = gym.spaces.Box(low=0, high=100000, shape=(6,), dtype=np.float32)

    def get_state(self):
        row = self.df.iloc[self.index]
        return np.array([row['CPU_Time(ms)'], row['Memory(KB)'], row['Energy(kWh)'],
                         row['Carbon_Emission(gCO₂)'], row['Loops'], row['Control_Flows']], dtype=np.float32)

    def step(self, action):
        reward = 0
        
        if action == self.df.iloc[self.index]['Efficiency_Label']:
            reward = 1
        else:
            reward = -1
    
        self.index += 1
        
        terminated = self.index >= len(self.df)  # True when all data points are processed
        truncated = False  # Modify this condition if you want an early termination
    
        if not terminated:
            state = self.get_state()
        else:
            state = None
        
        return state, reward, terminated, truncated, {}  # Return five values


    def reset(self, seed=None):
        """
        Reset the environment to the initial state.
        
        Args:
            seed (int, optional): Seed for the random number generator. Not used in this env.
        
        Returns:
            tuple: (observation, info) where observation is the initial state and info is an empty dict.
        """
        self.index = 0
        return self.get_state(), {}
```

```python
env = CodeEfficiencyEnv(df)
state = env.reset()
done = False

while not done:
    action = np.random.choice([0, 1])  # Random action
    state, reward, terminated, truncated, _ = env.step(action)  # Expect 5 values
    done = terminated or truncated  # End loop if either condition is met
    print(f"Action: {action}, Reward: {reward}")

```

```python
import gymnasium as gym  # Use gymnasium instead of gym
import numpy as np
import pandas as pd
from stable_baselines3 import DQN  # Deep Q-Learning Algorithm
from stable_baselines3.common.env_checker import check_env
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print(f"Training Samples: {len(train_df)}")
print(f"Testing Samples: {len(test_df)}")
```

```python
env = CodeEfficiencyEnv(train_df)
check_env(env)  # Optional: To verify the environment is built correctly
```

```python
model = DQN("MlpPolicy", env, verbose=1, learning_rate=0.001, gamma=0.99)
model.learn(total_timesteps=100000)
model.save("eco_code_ai_model")
print("Training Completed ✅")
```

```python
test_env = CodeEfficiencyEnv(test_df)
state, _ = test_env.reset()  # Unpack the tuple correctly
done = False
total_reward = 0

while not done:
    action, _ = model.predict(state)  # Now `state` is correctly formatted
    state, reward, done, _, _ = test_env.step(action)  # Unpack correctly
    total_reward += reward

print(f"Total Reward: {total_reward}")
```

```python
accuracy = (total_reward + len(test_df)) / (2 * len(test_df))
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```
